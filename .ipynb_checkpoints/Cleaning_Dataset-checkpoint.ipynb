```python
import pandas as pd
```

```python
df= pd.read_csv("./Resources/1976-2016-president.csv")
df.head()
```

```python
df.candidate = df.candidate.replace(
                                                                    {
                                                                        "Blank Vote/Scattering":"Void Vote",
                                                                        "Blank Vote/Void Vote/Scattering":"Void Vote",
                                                                        "Hedges, Jim":"Hedges, James",
                                                                        "Hoefling, Tom":"Hoefling, Thomas Conrad ""Tom""",
                                                                        "None Of The Above":"None Of These Candidates",
                                                                        "Scattering":"Void Vote",
                                                                        "White, Jerry":"White, Jerome ""Jerry"""
                                                                    })
```

```python
#keep data from 2012 and 2016 elections
df=df[df['year']>=2012]
df.head()
```

```python
#get all column names
for col in df.columns: 
    print(col) 
```

```python
#specify cols we want to keep and filter the df
cols_list=['year', 'state', 'state_po', 'candidate', 'party', 'writein', 'candidatevotes', 'totalvotes']
df=df[cols_list]
df.head()
```

```python
df['party']=df['party'].str.title()
df.head()
```

```python
#export cleaned csv
df.to_csv('./Resources/cleaned_data.csv', index=False)
```
